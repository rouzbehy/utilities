```python
pp_spec = read_csv("../../calcs/pp_2p76/jet_spec.csv",comment='#')
pp_spec['pT'] = 0.5*(pp_spec['pTmax']+pp_spec['pTmin'])
```

```python
## List of Kappas:
kappas = {0 :(1.0,1.0),
          1 :(1.0,15.0),
          2 :(15.0,1.0),
          3 :(15.0,15.0),
          4 :(14.4,8.2),
          5 :(7.4,1.2),
          6 :(10.9,11.7),
          7 :(3.9,4.7),
          8 :(5.6,13.4),
          9 :(12.6,6.4),
          10:(9.1,9.9),
          11:(2.1,2.9),
          12:(3.0,12.6),
          13:(10.0,5.6),
          14:(13.5,9.1),
          15:(6.5,2.1),
          16:(4.7,10.8),
          17:(11.7,3.8),
          18:(8.2,14.3),
          19:(1.2,7.3),
          20:(1.7,14.8),
          21:(8.7,7.8),
          22:(12.2,11.3),
          23:(5.2,4.3),
          24:(6.9,9.5),
          25:(13.9,2.5),
          26:(10.4,13.0),
          27:(3.4,6.0),
          28:(2.5,10.4),
          29:(9.5,3.4),
          30:(13.0,13.9),
          31:(6.0,6.9),
          32:(4.3,12.1),
          33:(11.3,5.1),
          34:(7.8,8.6),
          35:(14.8,1.6),
          36:(1.0,11.9),
          37:(8.0,4.9),
          38:(11.5,8.4),
          39:(4.5,1.4),
          40:(6.3,10.2),
          41:(13.3,3.2),
          42:(9.8,13.7),
          43:(2.8,6.7),
          44:(3.6,9.3),
          45:(10.6,2.3),
          46:(14.1,12.8),
          47:(7.1,5.8),
          48:(5.4,14.5),
          49:(12.4,7.5)}
all_kappas = [val for (k,val) in kappas.items()]
kappas_r, kappas_e = [e[0] for e in all_kappas],[e[1] for e in all_kappas]
```

```python
plt.scatter(kappas_r, kappas_e, color='gray')
linestyles = {1:'solid',2:'dotted',3:'dashed'}
rsets = [1]
kappa_sets = arange(0,50,2)
color = cm.Spectral(linspace(0, 1, len(kappa_sets)))
labels=[]
for (ik, kset) in enumerate(kappa_sets):
    (rad,elas) = kappas[kset]
    plt.scatter(rad,elas,color=color[ik], marker='P', s=200)
    labels.append(Line2D([],[],color=color[ik],label=f'{kset:d}:({rad:0.2f},{elas:0.2f})'))
plt.xlabel(r'$\kappa_{r}$')
plt.ylabel(r'$\kappa_{e}$')
```

```python
martini_data = {}
for rset in rsets:
    for kset in kappa_sets:
        fname = f"../../calcs/pbpb_2p76/param_fit_runs/rset_{rset}/kappaset_{kset}/jet_spectra.csv"
        martini_data[(rset,kset)] = read_csv(fname,comment='#')
```

```python
jet_fig, jet_ax = plt.subplots(3,1,sharex=True,sharey=True,figsize=(16,9))
jet_handles = []
jet_ax[0].set_title(f"Jet RAA")

for (iR, R) in enumerate([2,3,4]):
    plot_expr_data_on_axis(jet_ax[iR], CMS_JETS_00_05[f'0p{R}'], 'o')
    
pTmin, pTmax = 25, 300
pp_spec = pp_spec[pp_spec['pT'].between(pTmin,pTmax)]
for ir, rset in enumerate(rsets):
    for ik, kset in enumerate(kappa_sets):
        aa_jets = martini_data[(rset,kset)]
        aa_jets['pT'] = 0.5*(aa_jets['pTmax']+aa_jets['pTmin'])
        aa_jets = aa_jets[aa_jets['pT'].between(pTmin,pTmax)]
        for icone, R in enumerate([2, 3, 4]):
            raa_jets  = aa_jets[f'N0p{R}']/pp_spec[f'N0p{R}']
            draa_jets = raa_jets*sqrt((aa_jets[f'dN0p{R}']/aa_jets[f'N0p{R}'])*\
                                      (aa_jets[f'dN0p{R}']/aa_jets[f'N0p{R}'])+\
                                      (pp_spec[f'dN0p{R}']/pp_spec[f'N0p{R}'])*\
                                      (pp_spec[f'dN0p{R}']/pp_spec[f'N0p{R}']))
                
            jet_ax[icone].plot(pp_spec['pT'],raa_jets, color=color[ik], \
                               linestyle=linestyles[ir+1])
            jet_ax[icone].fill_between(pp_spec['pT'], raa_jets-draa_jets, \
                                              raa_jets+draa_jets, \
                                       color=color[ik],alpha=0.2)

jet_ax[0].set_ylim(top=1.5, bottom=-0.01)
jet_ax[0].set_xlim(right=350, left=60)
jet_ax[2].set_xlabel(r'$p^{\mathrm{jet}}_T$ (GeV)')
for icone, R in enumerate([0.2,0.3,0.4]):
    jet_ax[icone].text(0.01,0.85,f"R={R:0.1f}",transform=jet_ax[icone].transAxes,)
jet_fig.legend(handles=labels,bbox_to_anchor=(1.01,0.9),loc='upper left',fontsize=18)
jet_fig.supylabel(r'$R^{\mathrm{jet}}_{\mathrm{AA}}$')
jet_ax[0].set_xscale('log')

# jet_fig.savefig("../../plots/Thesis_Plots/RAW_JET_RAA_design_kappas.png",dpi=100)
```
