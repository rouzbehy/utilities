```python
def calc_Xi_sq(data, f_th, df_th, xmin=10, xmax=9999, include_theory_err=False):
    data = data[data['x'].between(xmin, xmax)]
    tx = data['x']
    ty = f_th(tx)
    dty = df_th(tx)
    chisq = array([(v1-v2)**2 for (v1,v2) in zip(ty,data['y'])])
    errs = data['dy_stat+']*data['dy_stat+'] + data['dy_syst+']*data['dy_syst+']
    if include_theory_err:
        errs = array([v1 + e*e for (v1,e) in zip(errs,dty)])
    chisq /= errs
    return sum(chisq)/len(data['y'])
```

```python
plt.scatter(kappas_r, kappas_e, color='gray')
linestyles = {1:'solid',2:'dotted',3:'dashed'}
rsets = [1]
kappa_sets = arange(0,50,2)
color = cm.Spectral(linspace(0, 1, len(kappa_sets)))
labels=[]
for (ik, kset) in enumerate(kappa_sets):
    (rad,elas) = kappas[kset]
    plt.scatter(rad,elas,color=color[ik], marker='P', s=200)
    labels.append(Line2D([],[],color=color[ik],label=f'{kset:d}:({rad:0.2f},{elas:0.2f})'))
plt.xlabel(r'$\kappa_{r}$')
plt.ylabel(r'$\kappa_{e}$')
```

```python
martini_data = {}
for rset in rsets:
    for kset in kappa_sets:
        fname = f"../../calcs/pbpb_2p76/param_fit_runs/rset_{rset}/kappaset_{kset}/hadron_spectra.csv"
        martini_data[(rset,kset)] = read_csv(fname,comment='#')
```

```python
had_fig, had_ax = plt.subplots(3,1,figsize=(16,9), sharex=True, sharey=True)
for (ir, rset) in enumerate(rsets):
    ax = had_ax[ir]
    plot_expr_data_on_axis(ax, CMS_HADS_00_05, 'o')
    plot_expr_data_on_axis(ax, ALICE_HADS_00_05, 'P')
    plot_expr_data_on_axis(ax, ATLAS_HADS_00_05, '*')
    for (ik, kset) in enumerate(kappa_sets):
        aa_spec = martini_data[(rset,kset)]
        raa = aa_spec['N']/pp_spec['N']
        draa = raa*sqrt(aa_spec['dN']*aa_spec['dN']/(aa_spec['N']*aa_spec['N'])+ \
                        pp_spec['dN']*pp_spec['dN']/(pp_spec['N']*pp_spec['N']))
        f_th = interp1d(pp_pT, raa)
        df_th = interp1d(pp_pT, draa)
        tmp=calc_Xi_sq(CMS_HADS_00_05, f_th,df_th,10, 9999, True)
        
        ax.plot(pp_spec['pT'],raa,color=color[ik], linestyle=linestyles[ir+1])
        ax.fill_between(pp_spec['pT'], raa+draa, raa-draa, color=color[ik], alpha=0.3)

had_ax[0].legend(handles=labels,bbox_to_anchor=(1.01,0.9),loc='upper left',fontsize=18)
had_ax[0].set_xlim(left=10, right=110)
had_ax[2].set_xlabel(r'$p^{h^{\pm}}_T$ (GeV)')
had_fig.supylabel(r'$R^{h^{\pm}}_{\mathrm{AA}}$')
#had_ax[0].set_xscale('log')
had_ax[0].set_ylim(top=1.0,bottom=0.11)

# had_fig.savefig("../../plots/Thesis_Plots/RAW_CHARGED_RAA_design_kappas.png",dpi=100)
```
